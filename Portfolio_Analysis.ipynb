```python
#lets build a portfolio which includes stock from
#STEEL: JSWSTEEL.NS
#ENERGY: TATAPOWER.NS
#AUTOMOBILE: TATAMOTORS.NS
#PHARMACEUTICAL SUNPHARMA.NS
```

```python
tickers=['JSWSTEEL.NS','SUNPHARMA.NS','TATAPOWER.NS','TATAMOTORS.NS']
start=dt.datetime(2019,1,1)
end=dt.date(2024,8,1)
```

```python
port=yp.download(tickers,start,end)
```

```python
port.columns
```

```python
port_1=port[['Adj Close']]
port_1
```

```python
port_1.plot(figsize=(25,18))
```

```python
#rate of growth of SHARES
port_growth=port_1/port_1.iloc[0]*100
port_growth.plot(figsize=(25,18))
```

```python
#return of portfolio per day
RETURNS_PORT=port_1/port_1.shift(1)-1
PER_DAY_RETURN=RETURNS_PORT.mean()
PER_DAY_RETURN
```

```python
#overall return for 5 years
overall_return=PER_DAY_RETURN*1377
overall_return
```

```python
#weightage of invetment per stock
weights=np.array([0.25,0.25,0.25,0.25])
```

```python
#overall average returns for 5 years
Average_returns_for_5_yrs=np.dot(overall_return,weights)
Average_returns_for_5_yrs
```

```python
print("average returns for your potfolio is:",Average_returns_for_5_yrs*100,'%')
```
